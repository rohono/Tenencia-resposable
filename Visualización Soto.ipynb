```python
# Instalando las bibliotecas
!pip install pandas matplotlib

# Importando las bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

# Cargando el archivo CSV en pandas
csv_path = '/content/drive/MyDrive/Colab Notebooks/base de datos proyecto/PROYECTO.2.csv'
df = pd.read_csv(csv_path, delimiter=';')

# Creando un gráfico de torta
plt.figure(figsize=(8, 8))
plt.pie(df['CANTIDAD'], labels=df['TENER MASCOTAS'], autopct='%1.1f%%', startangle=140)
plt.title('Encuesta El Chile que viene: Personas con y sin Mascotas')

# Muestra el gráfico
plt.show()
```
