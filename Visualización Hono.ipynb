```python
import matplotlib.pyplot as plt
import pandas as pd

# mi ruta
data = pd.read_csv('/content/PROYECTO.1.csv', delimiter=';')

# Definir categorías (las mismas que las del csv)
categorias = data['MASCOTAS']
registros = data['REGISTRO']
zona_urbana = data['zona urbana']
zona_rural = data['zona rural']

# Crear el gráfico, en mi caso, de barras
plt.bar(categorias, zona_urbana, label='Zona Urbana', color='aqua')
plt.bar(categorias, zona_rural, bottom=zona_urbana, label='Zona Rural', color='magenta')

# Datos de abajo
plt.xlabel('Mascotas')
plt.ylabel('Número de registros')
plt.title('Registros de Mascotas por Zona')
plt.legend()

# Imprimir
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Crear un DataFrame (esto me costó mucho entenderlo, finalmente no copié la ruta porque no pude y terminé escribiendo los datos)
data = {
    'Motivos': ['ESPACIO INSUFICIENTE', 'POCO TIEMPO', 'MUCHA RESPONSABILIDAD'],
    'Cantidad': [129019, 9353, 8386]
}

df = pd.DataFrame(data)

# Crear el gráfico de torta
plt.pie(df['Cantidad'], labels=df['Motivos'], autopct='%1.1f%%', startangle=90, colors=['magenta', 'aqua', 'yellow'])

plt.axis('equal')
plt.title('Motivos para no tener mascotas')

# Imprimir
plt.show()
```
